```python
import os
import cv2
import mediapipe as mp
import numpy as np
from ultralytics import YOLO  # Assuming YOLOv8 is being used

# Initialize Mediapipe pose and YOLO model
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True, model_complexity=1,
                    min_detection_confidence=0.7, min_tracking_confidence=0.7)


model = YOLO('runs/detect/gun-detection5/weights/best.pt')

# Path to test images
test_dir = '/content/drive/My Drive/gun-roboflow/test_set'
output_dir = '/content/drive/My Drive/gun-roboflow/output'  # Directory to save annotated images
os.makedirs(output_dir, exist_ok=True)  # Create output directory if it doesn't exist
counter = 0

# Iterate through test images
for img_name in os.listdir(test_dir):
    img_path = os.path.join(test_dir, img_name)
    counter += 1

    # Read the image
    image = cv2.imread(img_path)
    image_height, image_width, _ = image.shape

    # Run YOLO inference
    results = model(img_path, conf=0.2)  # Lower confidence threshold for partial detections

    # Extract YOLO detections
    detections = results[0].boxes


    class_ids = detections.cls.cpu().numpy().astype(int)
    bounding_boxes = detections.xyxy.cpu().numpy().astype(int)

    # Separate persons and guns
    persons = bounding_boxes[class_ids == 1]
    guns = bounding_boxes[class_ids == 0]

    # Process each detected person
    for person in persons:
        x_min, y_min, x_max, y_max = person

        # Expand the bounding box for better coverage
        padding = 20
        x_min = max(0, x_min - padding)
        y_min = max(0, y_min - padding)
        x_max = min(image_width, x_max + padding)
        y_max = min(image_height, y_max + padding)

        # Crop the person region
        person_crop = image[y_min:y_max, x_min:x_max]

        # Convert the cropped region to RGB for MediaPipe
        person_rgb = cv2.cvtColor(person_crop, cv2.COLOR_BGR2RGB)

        # Run pose estimation
        pose_results = pose.process(person_rgb)

        hand_positions = []

        if pose_results.pose_landmarks:
            for idx, landmark in enumerate(pose_results.pose_landmarks.landmark):
                if idx in [15, 16]:  # Right hand (16) and left hand (15) landmarks
                    x = int(landmark.x * (x_max - x_min) + x_min)
                    y = int(landmark.y * (y_max - y_min) + y_min)
                    hand_positions.append((x, y))
                    cv2.circle(image, (x, y), 5, (0, 255, 0), -1)  # Draw hand keypoints
        else:
            # Fallback for approximate hand positions if landmarks are missing
            hand_positions = [
                (x_min + (x_max - x_min) // 4, y_min),  # Left hand (approx)
                (x_min + 3 * (x_max - x_min) // 4, y_min)  # Right hand (approx)
            ]

        # Analyze hand proximity to guns
        for gun in guns:
            gx_min, gy_min, gx_max, gy_max = gun
            gun_center = ((gx_min + gx_max) // 2, (gy_min + gy_max) // 2)

            # Check if any hand is near the gun
            for hand in hand_positions:
                hx, hy = hand
                distance = ((hx - gun_center[0]) ** 2 + (hy - gun_center[1]) ** 2) ** 0.5
                distance_threshold = max((gx_max - gx_min) * 0.5, 50)  # Dynamic threshold

                if distance < distance_threshold:  # Threat detected
                    cv2.putText(image, "Threat Detected!", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
                    cv2.rectangle(image, (gx_min, gy_min), (gx_max, gy_max), (0, 0, 255), 2)  # Red box for guns
                    break

    # Save the annotated image to the output directory
    output_path = os.path.join(output_dir, img_name)
    cv2.imwrite(output_path, image)

print(f"Processed {counter} images and saved to {output_dir}.")
```
